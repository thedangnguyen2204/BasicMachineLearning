```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
```

```python
# Generate a dataset where the variations cannot be captured by a straight line
np.random.seed(0)
x, y = make_circles(n_samples=400, factor=.2, noise=0.02)
```

```python
plt.figure(1)
plt.title('Original Space')
plt.scatter(x[:,0], x[:,1], c = y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
# _ in labels mean subscript
```

```python
pca = PCA(n_components=2)
pca.fit(x)
x_pca = pca.transform(x)
```

```python
plt.figure(2)
plt.title('PCA')
plt.scatter(x_pca[:,0], x_pca[:,1], c = y)
plt.xlabel("$Component_1$")
plt.ylabel("$Component_2$")
# the components of PCA are unable to distinguish between the two classes in a linear fashion.
```

```python
class_1_indx = np.where(y==0)[0]
class_2_indx = np.where(y==1)[0]
class_2_indx
```

```python
x_pca
```

```python
x_pca[[0,1,2],1]
```

```python
plt.figure(3)
plt.title("PCA One Component")
plt.scatter(x_pca[class_1_indx, 0], np.zeros(len(class_1_indx)), color='red')
plt.scatter(x_pca[class_2_indx, 0], np.zeros(len(class_2_indx)), color='blue')
# The following graph,
# where we have plotted only the first component, explains how PCA is unable to differentiate the data
```

```python
# Using Kernel PCA
kpca = KernelPCA(kernel='rbf', gamma=10)
kpca.fit(x)
x_kpca = kpca.transform(x)
```

```python
plt.figure(4)
plt.title("Kernel PCA")
plt.scatter(x_kpca[:,0],x_kpca[:,1],c=y)
plt.xlabel("$Component_1$")
plt.ylabel("$Component_2$")
plt.show()
```
